```python
import pandas_datareader as web
import datetime as dt

start = dt.datetime(2022,1,1)
end = dt.datetime.now()

ltc = web.DataReader('BTC-USD', 'yahoo', start, end)
ltc
```

```python
import plotly.express as px

fig = px.line(ltc, y="High")
fig.show()
```
